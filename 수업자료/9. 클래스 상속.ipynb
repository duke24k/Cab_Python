# 9. 클래스 상속

## 상속

> 여러분이 클래스를 설계할때 기존에 이미 설계된 클래스와 유사하게 간다면 굳이 모든 내용을 전부 다시 짤필요는 없어요. 이미 설계된 클래스(부모클래스)로 부터 상속을 활용하면 상속받은 자식클래스에서는 부모클래스에 모든 것을 재사용 할수가 있답니다

상속을 받는 방법은 간단해요 `( )` 안에 상속받을 클래스의 이름을 적어주면 되요
```python
class 자식클래스(부모클래스):
    ...
```
이렇게 하면 부모클래스에 있는 모는 필드(field)과 함수(method)를 사용할 수 있어요

````python
class Coordinate:
    def __init__(self, x, y):
        self._x = x
        self._y = y
        
    @property
    def x(self):
        return self._x
    @x.setter
    def x(self, new_x):
        if type(new_x)!=float and type(new_x)!=int:
            raise ValueError("'{}' object is not available".format(type(new_x).__name__))
        self._x = new_x
        
    @property
    def y(self):
        return self._y
    @y.setter
    def y(self, new_y):
        if type(new_y)!=float and type(new_x)!=int:
            raise ValueError("'{}' object is not available".format(type(new_y).__name__))
        self._y = new_y
        
    def __add__(self, other):
        return Coordinate(self.x+other.x, self.y+other.y)
    def __sub__(self, other):
        return Coordinate(self.x-other.x, self.y-other.y)
    def __mul__(self, other):
        return self.x*other.x + self.y*other.y
    
    def __repr__(self):
        return "Coordinate({}, {})".format(self.x, self.y)
````

````python
point1 = Coordinate(1, 3)
point1
````

````python
point2 = Coordinate(2, 5)
point1 + point2
````

````python
class Coordinate_3d(Coordinate):
    def __init__(self, x, y, z):
        self._x = x
        self._y = y
        self._z = z
        
    @property
    def z(self):
        return self._z
    @z.setter
    def z(self, new_z):
        if type(new_z)!=float and type(new_x)!=int:
            raise ValueError("'{}' object is not available".format(type(new_z).__name__))
        self._z = new_z
        
    def __add__(self, other):
        return Coordinate_3d(self.x+other.x, self.y+other.y, self.z+other.z)
    def __sub__(self, other):
        return Coordinate_3d(self.x-other.x, self.y-other.y, self.z-other.z)
    def __mul__(self, other):
        return self.x*other.x + self.y*other.y + self.z*other.z
    
    def __repr__(self):
        return "Coordinate_3d({}, {}, {})".format(self.x, self.y, self.z)
````

````python
point1_3d = Coordinate_3d(1, 2, 3)
point1_3d
````

````python
point2_3d = Coordinate_3d(2, 2, 5)
point2_3d
````

````python
point1_3d + point2_3d
````
