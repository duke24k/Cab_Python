```python
from numpy import linspace, pi, cos
import matplotlib.pyplot as plt
X= linspace(-pi,pi,256)
C = cos(X)
plt.title('x Axis & y Axis with tick labels')
plt.plot(X, C)
plt.grid(True)

plt.xticks([-pi,-pi /2, 0, pi/2 ,pi],
         [r'$-\pi$',r'$-\pi/2$',r'$0$',r'$+\pi/2$',r'$+\pi$'])
plt.yticks([-1,0,1],['low,zero,high'])

plt.show()
```

```python
from numpy import linspace, pi, cos,sin,arange
import matplotlib.pyplot as plt
X= linspace(-pi,pi,256)
C ,S= cos(X),sin(X)
plt.title('plt with legends and axis title')
plt.plot(X,C, ls='--',label= 'cos')
plt.plot(X,S, ls=':',label= 'sin')
plt.legend(loc=2)
plt.ylabel('amplitude')
plt.xlabel('time')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x1 = np.linspace(0.,5.)
x2 = np.linspace(0.,2.)
y1 = np.cos(2 *np.pi *x1) * np.exp(-x1)
y2 = np.cos(2 *np.pi *x2) 

ax1 = plt.subplot(2,1,1)
plt.plot(x1,y1,'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped osciliation')
print(ax1)

ax2 = plt.subplot(2,1,2)
plt.plot(x2,y2,'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
print(ax2)

plt.tight_layout()
plt.show()
```

```python
from numpy import linspace, cos, pi
import matplotlib.pyplot as plt
x= linspace(0,2.0*pi,40)
y= cos(x)
plt.plot(x,y,'k-')
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = cos(x)')
plt.show()
```

```python
from numpy import linspace,sin, cos, pi
import matplotlib.pyplot as plt
x= linspace(0,2.0*pi,40)
yc = cos(x)
ys = sin(x)
plt.plot(x,yc,'ko-',label = 'cos(x)')
plt.plot(x,ys,'k--',label = 'sin(x)')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = cos(x) and sin(x)')
plt.show()
```
