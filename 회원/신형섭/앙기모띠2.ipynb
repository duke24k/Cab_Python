```python
import matplotlib.pyplot as plt
from math import exp
import numpy as np

def fac(n):
    if(n==1):
        return 1;
    return fac(n-1)*n

def p4_0(x):
    return 1 + x + x**2/fac(2) + x**3/fac(3) + x**4/fac(4)
def p4_1(x):
    return exp(1)*(1 + (x-1) + (x-1)**2/fac(2) + (x-1)**3/fac(3) + (x-1)**4/fac(4))
x = np.arange(0,3,0.1)
y1 = [p4_0(v) for v in x]
y2 = [p4_1(v) for v in x]
plt.plot(x,y1,c='b')
plt.plot(x,y2,c='r')
y3 = [abs(exp(v)-p4_0(v)) for v in x]
y4 = [abs(exp(v)-p4_1(v)) for v in x]
plt.plot(x,y3,c='b')
plt.plot(x,y4,c='r')
ans = [exp(v) for v in x]
plt.plot(x,ans,c='k')
plt.show()
# 크기가 커질수록 x가 1일때의 테일러 급수의 근사값이 오차가 적다
```

```python
import matplotlib.pyplot as plt
from math import exp
import numpy as np

def fac(n):
    if(n==1):
        return 1;
    return fac(n-1)*n

def p4_0(x):
    return 1 + x + x**2/fac(2) + x**3/fac(3) + x**4/fac(4)
def p4_1(x):
    return exp(1)*(1 + (x-1) + (x-1)**2/fac(2) + (x-1)**3/fac(3) + (x-1)**4/fac(4))
x = np.arange(0,0.01,0.001)
# 테일러 급수
y1 = [p4_0(v) for v in x]
y2 = [p4_1(v) for v in x]
plt.plot(x,y1,c='b')
plt.plot(x,y2,c='r')
# 오차식
y3 = [abs(exp(v)-p4_0(v)) for v in x]
y4 = [abs(exp(v)-p4_1(v)) for v in x]
plt.plot(x,y3,c='b')
plt.plot(x,y4,c='r')
ans = [exp(v) for v in x]
plt.plot(x,ans,c='k')
plt.ylim(0.99,1.03)
plt.show()
# x가 0에 가까울때는 p4_0이 오차가 더 적다
```
